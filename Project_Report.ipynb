## Abstract

This project explores the predictive value of insider trading behavior by analyzing Form 4 filings from the U.S. Securities and Exchange Commission (SEC) and corresponding stock price movements. It is supposed to give an overview on the methods and thought process that resulted in an algorithmic trading stratagy in form of a report. 
The central hypothesis is ##unfertig## that repeat purchases of a company's stock by the same insider—especially within short timeframes—carry informational value that may precede future price movements.  

##unfertig##To investigate this, I engineered a set of event-based features capturing historical insider activity (e.g., prior purchases within 7 days) and combined them with market data from Yahoo Finance.  
Using exploratory data analysis and classification models, I examined the relationship between insider activity patterns and short-term stock performance.  

##unfertig##The results suggest that clustered insider purchases, particularly by the same individual, can provide meaningful signals in specific market contexts.  
This report outlines the full analytical process, from data collection and cleaning to feature engineering, modeling, and interpretation.
