```python
import pandas as pd
insider_data = pd.read_csv('data/close_data_2018-01.csv')

insider_data["prior_filling_of_same_person"] = (
    insider_data
    .groupby(["reportingOwner_name", "issuer_tradingSymbol"])
    .cumcount() > 0
)

insider_data['others_fillings_of_same_stock'] = (
    insider_data
    .groupby(["issuer_tradingSymbol"])
    .cumcount() > 0
)

insider_data['amount_fillings_of_same_person'] = (
    insider_data
    .groupby(["issuer_tradingSymbol"])
    .cumcount()
)

# sicherstellen, dass Datum datetime ist
insider_data["transactionDate"] = pd.to_datetime(insider_data["transactionDate"])

# leere Spalte initialisieren
insider_data["purchases_last_7d"] = 0

# durch jede Zeile iterieren (langsamer, aber einfach)
for idx, row in insider_data.iterrows():
    name = row["reportingOwner_name"]
    ticker = row["issuer_tradingSymbol"]
    date = row["transactionDate"]
    
    # alle früheren Trades derselben Person im selben Ticker in den letzten 7 Tagen
    mask = (
        (insider_data["reportingOwner_name"] == name) &
        (insider_data["issuer_tradingSymbol"] == ticker) &
        (insider_data["transactionDate"] < date) &
        (insider_data["transactionDate"] >= date - pd.Timedelta(days=7))
    )
    
    insider_data.at[idx, "fillings_last_7d"] = mask.sum()

# Binary-Flag
insider_data["has_prior_fillings_7d"] = insider_data["fillings_last_7d"] > 0
```
